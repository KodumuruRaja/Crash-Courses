```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
x=data.drop(['tip'],axis=1)
y=data['tip']
```

```python
x.head()
```

```python
y.head()
```

```python
data.info()
```

```python
x['day']=x['day'].cat.codes
x
```

```python
x['sex']=x['sex'].cat.codes
x
```

```python
x['smoker']=x['smoker'].cat.codes
x
```

```python
x['time']=x['time'].cat.codes
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=200).fit(x_train,y_train)
```

```python
from shapash.explainer.smart_explainer import SmartExplainer
```

```python
se=SmartExplainer()
```

```python
se.compile(x=x_test,model=reg)
```

```python
se
```

```python
app=se.run_app(title_story='Tips Dataset')
```

```python
predict=se.to_smartpredictor()
predict.save('./predict.pkl')
```

```python
from shapash.utils.load_smartpredictor import load_smartpredictor
predict_load=load_smartpredictor('./predict.pkl')
```

```python
predict_load.add_input(x=x,ypred=y)
```
