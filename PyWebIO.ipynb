```python
from pywebio.output import * # Displaying Alert Mesage and displaying Text along with Txt File available to download locally
# Text Output
put_text("Kodumuru Raja")

# Table Output
put_table([
    ['Name', 'RollNo'],
    ['Raja', '15'],
    ['Nivi', '16'],
    ['Rishi', '17']
])

# Markdown Output
put_markdown('~~Vittal~~')

# File Output
put_file('Kodumuru.txt', b'Raja Kodumuru')

# PopUp Output
popup('Alert', 'You can make this without using HTML and JS')
```
